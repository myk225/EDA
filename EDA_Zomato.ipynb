```python
top10_cuisines = df["Cuisines"].value_counts().head(10)
print(top10_cuisines)
```

```python
# Split cuisines by comma, stack them into a single series, then count
top10_cuisines = (
    df["Cuisines"]
    .dropna()
    .str.split(",")          # split by comma
    .explode()               # turn lists into rows
    .str.strip()             # remove extra spaces
    .value_counts()
    .head(10)
)

print(top10_cuisines)
```
